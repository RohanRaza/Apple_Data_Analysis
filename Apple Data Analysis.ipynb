# Avg User Rating of Free Games

# Avg User Rating of Free Games And Music

```python
free_game_music_rating = []
for i in data[1:]:
    price = float(i[4])
    user_rating = float(i[7])
    if price == 0 and (i[11] =="Games" or i[11] == "Music"):
        free_game_music_rating.append(user_rating)
    
```

```python
round(sum(free_game_music_rating)/len(free_game_music_rating),2)
```

# Frequency Table of Prime Genre

```python
data[0].index("prime_genre")
```

```python
prime_genre_frequeny = {}
for i in data[1:]:
    prime = i[11]
    if prime in prime_genre_frequeny:
        prime_genre_frequeny[prime] += 1
    else:
        prime_genre_frequeny[prime] = 1
```

```python
prime_genre_frequeny
```

# Frequency Table of Free Prime Genre

```python
free_prime_genre = {}
for i in data[1:]:
    price = float(i[4])
    prime = i[11]
    if price == 0:
        if prime in free_prime_genre:
            free_prime_genre[prime] += 1
        else:
            free_prime_genre[prime] = 1
```

```python
free_prime_genre
```

# Frequency Table of Paid Prime Genre

```python
paid_prime_genre = {}
for i in data[1:]:
    price = float(i[4])
    prime = i[11]
    rating = float(i[7])
    if price > 0 and rating > 4:
        if prime in paid_prime_genre:
            paid_prime_genre[prime] += 1
        else:
            paid_prime_genre[prime] = 1
    
```
